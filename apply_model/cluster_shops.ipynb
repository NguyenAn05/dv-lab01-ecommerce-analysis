```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# Đọc dữ liệu từ file shops.csv
df_shops = pd.read_csv('../data/cleaned/shops.csv')

# Lọc bỏ các dòng bị thiếu dữ liệu ở 3 cột quan trọng (nếu có)
df_shops = df_shops.dropna(subset=['follower_count', 'rating_star', 'response_rate'])

# Chọn các biến đưa vào mô hình
features = ['follower_count', 'rating_star', 'response_rate']
X = df_shops[features]

# Chuẩn hóa dữ liệu để đưa các biến về cùng một hệ quy chiếu
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Khởi tạo và chạy mô hình K-Means với 3 cụm
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df_shops['Cluster_ID'] = kmeans.fit_predict(X_scaled)
```

```python
# Phân tích đặc điểm của từng cụm để tiến hành Gắn nhãn
# Tính giá trị trung bình của 3 chỉ số theo từng cụm
cluster_summary = df_shops.groupby('Cluster_ID')[features].mean().round(2)
print("--- TRUNG BÌNH CÁC CHỈ SỐ THEO TỪNG CỤM ---")
print(cluster_summary)
```

```python
label_mapping = {
    0: 'Dẫn đầu', 
    1: 'Tiềm năng',
    2: 'Báo động',
}

# Tạo cột nhãn chính thức
df_shops['Cluster_Label'] = df_shops['Cluster_ID'].map(label_mapping)

df_shops.drop(columns=['Cluster_ID'], inplace=True)

# Xuất file dữ liệu mới để mang lên Dashboard (Tableau/PowerBI)
df_shops.to_csv('../data/cleaned/shops_clustered.csv', index=False, encoding='utf-8-sig')
print("Đã lưu kết quả phân cụm vào file 'shops_clustered.csv'!")
```
